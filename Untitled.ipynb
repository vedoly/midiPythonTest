```python
import py_midicsv
import numpy as np
import pandas as pd
csv_a = py_midicsv.midi_to_csv("song.mid")
csv_b = py_midicsv.midi_to_csv("song2.mid")

Ans=[]
Ans2=[]
for k in csv_a:
    L=k.strip().split(", ")
    if L[2] == "Note_on_c":
        L[2] = 1
        Ans.append(L)
    if L[2] == "Note_off_c":
        L[2] = 0
        Ans.append(L)

Ans 
        
```
